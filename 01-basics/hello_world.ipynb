```python
integer = 1
float = 1.0
string = "Hello, World!"
boolean = True
print(integer, float, string, boolean)
print(type(integer), type(float), type(string), type(boolean))
```

# Variable

- 意義：變數是程式中用來儲存資料的命名位置。
- 例如：
  - age = 20 -> 表示我們將 20 這個值儲存在 age 這個變數中
  - name = "John" -> 表示我們將 "John" 這個值儲存在 name 這個變數中
- 命名規則：
  - 只能包含字母、數字和底線。
  - 不能以數字開頭。
  - 不能使用 Python 保留字，如 if, else, for, while, def, return, etc.
  - 慣例上使用小寫字母，並以底線分隔單字。

```python
age = 20
name = "John"
print(name)
print(age)
print(name, age)
print(f"Hello, {name}! You are {age} years old.")
```

# Expression - Basic Arithmetic Operations

```python
print(1 + 1)
print(1 - 1)
print(3 * 4)
print(4 / 2)
print(7 % 4) # 餘數運算
print(2 ** 3) # 指數運算
print(7 // 3) # 整數除法
```

# Data Structures - List, Tuple, Dictionary, Set

```python
list = [1, 2, 3, 4, 5]
tuple = (1, 2, 3, 4, 5)
dictionary = {"name": "John", "age": 20}
set = {1, 2, 3, 4, 5}
print(list, tuple, dictionary, set)
print(type(list), type(tuple), type(dictionary), type(set))
```

# Condition - if, else

```python
if age > 18:
    print("You are an adult")
elif age == 18:
    print("You are a teenager")
else:
    print("You are a minor")
```
