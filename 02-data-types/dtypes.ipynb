### **List operations** â­ï¸â­ï¸â­ï¸â­ï¸â­ï¸
##### [element1, element2, element3, ....]
- **Mutable**
- Common uses: Storing sequences, iteration, sorting

````python
members = ["Darrius", "Tyson", "HonYi", "YoNing", "Zhu"]
drinks = ["ç„¦ç³–ç‘ªå…¶æœµ", "ç¦å‰èŒ¶é‚£æ", "æ˜Ÿå†°æ¨‚", "é†‡æ¿ƒæŠ¹èŒ¶é‚£æ", "å†°æ–èŒ¶"]
numbers = [1, 2, 3, 4, 5]
````

##### GET / SLICE VALUE(S): list_[n] & list_[i:j]

````python
members[0]
print("members[0] ->", members[0])
members[0:2]
print("members[0:2] ->", members[0:2])
members[:2]
print("members[:2] ->", members[:2])
members[2:]
print("members[2:] ->", members[2:])
members[-1]
print("members[-1] ->", members[-1])
members[-3:-1]
print("members[-3:-1] ->", members[-3:-1])
````

##### ADD VALUES: append() & insert() & extend()

````python
# append & insert & extend
all_data = []

members.append("WJ")
print("members.append('WJ') ->", members)
members.insert(0, "WJ")
print("members.insert(0, 'WJ') ->", members)
all_data.extend(drinks)
print("members.extend(drinks) ->", all_data)
all_data = members + drinks
print("members + drinks ->", all_data)
````

##### REMOVE VALUE(S): remove() & pop() & clear()

````python
# remove & pop
members.remove("WJ") # remove the first occurrence of the value
print("members.remove('WJ') ->", members)
members.pop() # remove the last element
print("members.pop() ->", members)
members.clear() # remove all elements
print("members.clear() ->", members)
````

##### CHANGE ORDERS: reverse() & sort()

````python
# reverse & sort
numbers.reverse()
print("numbers.reverse() ->", numbers)
numbers.sort() # sort the list in ascending order
print("numbers.sort() ->", numbers)
numbers.sort(reverse=True) # sort the list in descending order
print("numbers.sort(reverse=True) ->", numbers)
````

##### COUNT VALUES: count() & len()

````python
# count how many times the value appears in the list
members.count("WJ")
print("members.count('WJ') ->", members.count("WJ"))

# count how many values in the list
len(members)
print("len(members) ->", len(members))
````

### Set operations â­ï¸â­ï¸â­ï¸
##### {element1, element2, ...}
- **Mutable**
- Common uses: **Removing duplicates**, set operations

````python
duplicates = [1, 2, 2, 3, 4, 4, 5]
duplicated_names = ["Darrius", "Tyson", "Darrius", "HonYi", "YoNing", "Darrius", "Zhu"]
````

````python
unique_numbers = set(duplicates)
print("unique_numbers ->", unique_numbers)

unique_names = list(set(duplicated_names))
print("unique_names ->", unique_names)
````

##### ADD VALUE(S): add()

````python
unique_numbers.add(6)
print("unique_numbers.add(6) ->", unique_numbers)
````

##### UPDATE VALUE(S): update()

````python
unique_numbers.update([7, 8, 9])
print("unique_numbers.update([7, 8, 9]) ->", unique_numbers)
````

##### REMOVE VALUE(S): remove() & discard()

````python
# remove the value if it exists (error if the value doesn't exist)
unique_numbers.remove(6)
print("unique_numbers.remove(6) ->", unique_numbers)

# remove the value if it exists (no error if the value doesn't exist)
unique_numbers.discard(6)
print("unique_numbers.discard(6) ->", unique_numbers)

##### CLEAR SET: clear()
unique_numbers.clear()
print("unique_numbers.clear() ->", unique_numbers)
````

# List / Dict / Set comprehension

Main purpose:
- Generate a new list / dict / set
- Cleaner code (Complete the loop + condition judgment + data structure creation in one line)

ðŸ”¹ Syntax:
```python
[expression for variable in iterable if condition]

````python
members = ["Darrius", "Tyson", "HonYi", "YoNing", "Zhu", "Zhu"]
drinks = ["ç„¦ç³–ç‘ªå…¶æœµ", "ç¦å‰èŒ¶é‚£æ", "æ˜Ÿå†°æ¨‚", "é†‡æ¿ƒæŠ¹èŒ¶é‚£æ", "å†°æ–èŒ¶", "å†°ç¾Žå¼"]
numbers = [1, 2, 3, 4, 5]

name_drink_pair = []
for member, drink in zip(members, drinks):
    if member == "Darrius":
        name_drink_pair.append((member, drink))
````

````python
# list comprehension
name_drink_pair = [(member, drink) for member, drink, number in zip(members, drinks, numbers) if number > 2]
print("name_drink_pair ->", name_drink_pair)
````

````python
# dict comprehension
name_drink_pair = {member: drink for member, drink in zip(members, drinks)} # é‡è¤‡çš„æœƒè¢«è¦†è“‹
print("name_drink_pair ->", name_drink_pair)
````

````python
# set comprehension
name_drink_pair = {member for member in members} # é‡è¤‡çš„æœƒè¢«åŽ»é™¤
print("name_drink_pair ->", name_drink_pair)
````

````python
for index, member in enumerate(members):
    print(f"index: {index}, member: {member}")

# éœ€è¦ç”¨ enumerate() ä¾†å–å¾— index çš„ case
# å»ºç«‹ç·¨è™Ÿåˆ°é£²æ–™çš„å°æ‡‰
print("=== é£²æ–™ç·¨è™Ÿå°ç…§è¡¨ ===")
for number, drink in enumerate(drinks, start=1111):
    print(f"{number}: {drink}")

# dict comprehension
drink_menu = {number: drink for number, drink in enumerate(drinks, start=1001)}
print("drink_menu ->", drink_menu)
````
