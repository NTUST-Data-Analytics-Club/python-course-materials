# Import necessary packages

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Load data

```python
spend_data = pd.read_csv('../data/sample.csv')
spend_data.head()
```

# Group by category and calculate total amount

```python
category_summary = spend_data.groupby(by="category", as_index=False)["amount"].sum()
category_summary = category_summary.sort_values(by="amount", ascending=False)
category_summary
```

# Group by date and calculate total amount

```python
date_summary = spend_data.groupby(by="date", as_index=False)["amount"].sum()
date_summary = date_summary.sort_values(by="amount", ascending=False)
date_summary
```

# Plot total amount by category

```python
category_summary.plot(kind="bar", x="category", y="amount", figsize=(10, 6))
```
